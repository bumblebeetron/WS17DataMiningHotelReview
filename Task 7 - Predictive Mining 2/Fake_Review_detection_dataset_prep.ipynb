```python
class fakeClassifier:
    def __init__(self):
        self.browser = Browser('chrome', headless=True)
    def connect(self):
        self.browser.visit('http://reviewskeptic.com/')
    def isAuthentic(self,text):
        #self.browser.visit('http://reviewskeptic.com/')
        self.browser.find_by_name("review_text").fill(text)
        self.browser.find_by_value("Test It").click()
        truthful = self.browser.find_by_text('Truthful')
        if not truthful:
            return 0
        else:
            return 1


        
```

```python
def from0to20k():
    classifier= fakeClassifier()
    classifier.connect()
    
    for i in range(0,18000):
        try:
            if(i%10==0):
                print("0th",i)
            if(i%500==0):
                    Hotel_Reviews_UK.to_csv("Hotel_Reviews_UK_0_20_fake.csv")
                    print("saving 0 to csv")
            if 'No Positive' not in Hotel_Reviews_UK["Positive_Review"].iloc[i]:
                Hotel_Reviews_UK.Authentic.iloc[i]=classifier.isAuthentic(Hotel_Reviews_UK["Positive_Review"].iloc[i])
                continue
            if 'No Negative' not in Hotel_Reviews_UK["Negative_Review"].iloc[i]:
                Hotel_Reviews_UK.Authentic.iloc[i]=classifier.isAuthentic(Hotel_Reviews_UK["Negative_Review"].iloc[i])
                continue
        except:
            print("error occured on 0 at index ",i)
            a=i
            break
```
