```python
#sentiment analysis
from textblob import TextBlob
```

```python
#sentiment analysis
HotelReviewsText = pd.read_csv('Hotel_Reviews.csv',usecols=['Positive_Review','Negative_Review','Reviewer_Score'])
HotelReviewsText["Positive_Polarity"]="-"
```

```python
#sentiment analysis
from textblob.sentiments import NaiveBayesAnalyzer
    



```

```python
#sentiment analysis
for i in HotelReviewsText.index:
    polarity=0
    if ("No Negative" not in HotelReviewsText['Negative_Review'][i]):
           polarity+=TextBlob(HotelReviewsText['Negative_Review'][i], analyzer=NaiveBayesAnalyzer()).sentiment.p_pos
    if ("No Positive" not in HotelReviewsText['Positive_Review'][i]):
           polarity+=TextBlob(HotelReviewsText['Positive_Review'][i], analyzer=NaiveBayesAnalyzer()).sentiment.p_pos
    HotelReviewsText["Positive_Polarity"][i]=polarity
```
