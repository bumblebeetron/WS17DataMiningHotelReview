```python
tagsDict=collections.defaultdict(set)
tagsList=[]

for rowNo in hotelReviewTags.index:
    row=hotelReviewTags['Tags'][rowNo].split(",")
    tags =[]
    for tag in row:
        tagValue= tag.strip('[').strip(']').strip(" '").strip("' ").lower()
        tags.append(tagValue)
        tagsDict[tagValue].add(rowNo)
    tagsList.append(tags)
    
```

```python
print(len(tagsDict['couple']))
```

```python
tagsFreq=collections.defaultdict(int)
for key in tagsDict:
    tagsFreq[key]= len(tagsDict[key])
    
        
        
     


 
```

```python
from mlxtend.preprocessing import OnehotTransactions
from mlxtend.frequent_patterns import apriori


```

```python
oht = OnehotTransactions()
oht_ary = oht.fit(tagsList).transform(tagsList)  
df = pd.DataFrame(oht_ary, columns=oht.columns_)  
df
```

```python
frequent_itemsets = apriori(df, min_support=0.05, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
def cnvListToStr(listObj):
    
    return (''.join(listObj))
```

```python
#market basket on core root words
exclusiveList=[]
for i in range(15):  
    for j in range(i+1, 15, 1):
        if (i>5):
            if not (tagsDict[cnvListToStr(frequent_itemsets['itemsets'][i])] & tagsDict[cnvListToStr(frequent_itemsets['itemsets'][j])]):
                exclusiveList.append([cnvListToStr(frequent_itemsets['itemsets'][i]),cnvListToStr(frequent_itemsets['itemsets'][j])])
                
print (exclusiveList)
#no abuse of tags  
```
