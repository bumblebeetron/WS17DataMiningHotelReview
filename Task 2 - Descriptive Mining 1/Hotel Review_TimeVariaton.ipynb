Overall over all hotels

```python
month_array=np.array([])
day_array=np.array([])
year_array=np.array([])
week_array=np.array([])

average_rating_monthly=np.array([])
average_rating_yearly=np.array([])
average_rating_daily=np.array([])
average_rating_weekly=np.array([])

visitor_monthly=np.array([])
visitor_yearly=np.array([])
visitor_daily=np.array([])
visitor_weekly=np.array([])
```

```python
all_months=np.sort(np.array(Hotel_Reviews.Month.unique()))
all_year=np.sort(np.array(Hotel_Reviews.Year.unique()))
all_days=np.sort(np.array(Hotel_Reviews.Day.unique()))
```

```python
all_daysweeks=np.sort(np.array(Hotel_Reviews.WeekDay.unique()))
```

## Analysis Month

```python
for month in all_months :
    TEMP=Hotel_Reviews.loc[Hotel_Reviews['Month']==month]
    
    month_array=np.append(month_array,month)
    
    average_rating_monthly=np.append(average_rating_monthly,TEMP['Reviewer_Score'].mean())
    
    visitor_monthly=np.append(visitor_monthly,TEMP.shape[0])
    
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.subplot(1, 2, 1)
plt.plot(month_array,average_rating_monthly)
plt.xlabel('Month')
plt.ylabel('Mean Rating')
plt.title("Monthly Rating Distubution")

plt.subplot(1, 2,2)
plt.plot(month_array,np.divide(np.subtract(average_rating_monthly,np.mean(average_rating_monthly)),np.std(average_rating_monthly)))
plt.xlabel('Month')
plt.ylabel('Mean Rating(Scaled)')
plt.title("Monthly Rating Distubution(Normalised)")
plt.tight_layout() 
plt.rcParams['figure.figsize'] = (11, 4)
```

```python
plt.subplot(1, 2, 1)
plt.plot(month_array,visitor_monthly)
plt.xlabel('Month')
plt.ylabel('Mean Visitor')
plt.title("Monthly Visitor Distubution")

plt.subplot(1, 2, 2)
plt.plot(month_array,np.divide(np.subtract(visitor_monthly,np.mean(visitor_monthly)),np.std(visitor_monthly)))
plt.xlabel('Month')
plt.ylabel('Mean Visitor(Normalised)')
plt.title("Monthly Visitor Distubution(Normalised)")
plt.tight_layout()
plt.rcParams['figure.figsize'] = (11, 4)
```

```python
Vistiors,=plt.plot(month_array,np.divide(np.subtract(visitor_monthly,np.mean(visitor_monthly)),np.std(visitor_monthly)),label='Visitor')
Rating,=plt.plot(month_array,np.divide(np.subtract(average_rating_monthly,np.mean(average_rating_monthly)),np.std(average_rating_monthly)),label='Rating')
plt.xlabel('Month')
plt.ylabel('Mean')
plt.legend(handles=[Vistiors,Rating])

plt.title("Monthly Scaled Visitor vs Rating")
plt.rcParams['figure.figsize'] = (11, 4)
```

## Correlation Finding Test

```python
Hotel_Reviews['Reviewer_Score_scaled']=(Hotel_Reviews['Reviewer_Score']-Hotel_Reviews['Reviewer_Score'].mean())/Hotel_Reviews['Reviewer_Score'].std()
Hotel_Reviews['WeekDay_scaled']=(Hotel_Reviews['WeekDay']-Hotel_Reviews['WeekDay'].mean())/Hotel_Reviews['WeekDay'].std()
Hotel_Reviews['Year_scaled']=(Hotel_Reviews['Year']-Hotel_Reviews['Year'].mean())/Hotel_Reviews['Year'].std()
Hotel_Reviews['Month_scaled']=(Hotel_Reviews['Month']-Hotel_Reviews['Month'].mean())/Hotel_Reviews['Month'].std()
Hotel_Reviews['Day_scaled']=(Hotel_Reviews['Day']-Hotel_Reviews['Day'].mean())/Hotel_Reviews['Day'].std()
```
