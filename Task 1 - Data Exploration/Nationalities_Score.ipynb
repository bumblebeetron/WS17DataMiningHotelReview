```python
import pandas as pd
import numpy as np
```

```python
Hotel_Reviews=pd.read_csv("Hotel_Reviews.csv")
Hotel_Reviews.head()
```

```python
Hotel_Reviews.shape
```

```python
Nationalities=Hotel_Reviews.Reviewer_Nationality.unique()
```

```python
Nationalities.size
```

```python
Reviewer_Nationality=np.array([])
mean_reviewer_rating=np.array([])
mode_reviewer_rating=np.array([])
std_reviewer_rating=np.array([])
outlier_threshold=np.array([])
inlier_threshold=np.array([])
mean_average_score=np.array([])
```

```python
for Nationality in Nationalities:
    TEMP=Hotel_Reviews.loc[Hotel_Reviews['Reviewer_Nationality']==Nationality]
    
    Reviewer_Nationality=np.append(Reviewer_Nationality,Nationality)
    
    mean_reviewer_rating=np.append(mean_reviewer_rating,TEMP['Reviewer_Score'].mean())
    mode_reviewer_rating=np.append(mode_reviewer_rating,TEMP['Reviewer_Score'].mode())
    std_reviewer_rating=np.append(std_reviewer_rating,TEMP['Reviewer_Score'].std())
    
    outlier_threshold=np.append(outlier_threshold,TEMP['Reviewer_Score'].mean()+(2*TEMP['Reviewer_Score'].std()))
    inlier_threshold=np.append(inlier_threshold,TEMP['Reviewer_Score'].std())
    
    mean_average_score=np.append(mean_average_score,TEMP['Average_Score'].mean())
```

```python
columns=['Reviewer_Nationality','mean_reviewer_rating','std_reviewer_rating','outlier_threshold','inlier_threshold','mean_average_score']
```

```python
data=np.column_stack((Reviewer_Nationality,mean_reviewer_rating,std_reviewer_rating,outlier_threshold,inlier_threshold,mean_average_score))
```

```python
HotelReviews=pd.DataFrame(data,columns=columns)
```
