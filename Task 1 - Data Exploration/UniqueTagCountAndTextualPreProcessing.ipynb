```python
hotelReviewTags = pd.read_csv('Hotel_Reviews.csv',usecols=['Tags'])
```

```python
uniqueTags = set()
for rowNo in hotelReviewTags.index:
    row=hotelReviewTags['Tags'][rowNo].split(",")
    for tag in row:
        uniqueTags.add(tag.strip('[').strip(']').strip(" '").strip("' "))

```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
HotelReviews = pd.read_csv('Hotel_Reviews.csv',usecols=['Positive_Review','Negative_Review'])
```

```python
corpus= []
uniqueUsefulTokens= set()

for i in range(10000):
       review=""
       if ("No Negative" in HotelReviews['Negative_Review'][i]) and "No Positive" in HotelReviews['Positive_Review'][i]:
            continue
       elif "No Negative" in HotelReviews['Negative_Review'][i]:
            review = HotelReviews['Positive_Review'][i]
       else:
            review = HotelReviews['Negative_Review'][i]
       
       review= re.sub('[^a-zA-Z]', ' ', review)
       review=review.lower().split()
       ps=PorterStemmer()
       review  = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
       for word in review:
            uniqueUsefulTokens.add(word)
       review = ' '.join(review)
       corpus.append(review)
```

```python
print (len(uniqueUsefulTokens), "unique value adding words after textual pre-processing, identified in the dataset in the first 10000 rows")
```
