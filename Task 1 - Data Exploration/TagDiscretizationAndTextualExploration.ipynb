```python
import langid
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
HotelReviews = pd.read_csv('Hotel_Reviews.csv',usecols=['Positive_Review','Negative_Review'])
```

```python
#count all unique languages used in valid reviews
languageCount= set()
for i in HotelReviews.index:
       review=""
       if ("No Negative" in HotelReviews['Negative_Review'][i]) and "No Positive" in HotelReviews['Positive_Review'][i]:
            continue
       elif "No Negative" in HotelReviews['Negative_Review'][i]:
            review = HotelReviews['Positive_Review'][i]
       else:
            review = HotelReviews['Negative_Review'][i]
       languageCount.add(langid.classify(review)[0])
```

```python
print("Approx. ", len(languageCount), " different languages across all reviews in the dataset.They are : ", languageCount)
```

```python
#code for selecting the row, and extracting the only the english Reviews( either positive or negative, whichever exists)
#then subjecting it to NLP pre-processing( special character/number removal, stop-word removal, stemming, tokenization),
#and storing the results in a unique corpus for each english review. also Adding all the unique words to a uniqueTokens list
corpus= []
uniqueUsefulTokens= set()

for i in HotelReviews.index:
       review=""
       if ("No Negative" in HotelReviews['Negative_Review'][i]) and "No Positive" in HotelReviews['Positive_Review'][i]:
            continue
       elif "No Negative" in HotelReviews['Negative_Review'][i]:
            review = HotelReviews['Positive_Review'][i]
       else:
            review = HotelReviews['Negative_Review'][i]
       
       if (langid.classify(review)[0]!='en'):
            continue
       review= re.sub('[^a-zA-Z]', ' ', review)
       review=review.lower().split()
       ps=PorterStemmer()
       review  = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
       for word in review:
            uniqueUsefulTokens.add(word)
       review = ' '.join(review)
       corpus.append(review)
```

```python
print (len(uniqueUsefulTokens), "unique value adding words after textual pre-processing of only english reviews in the dataset(all other languages were skipped)")
```
