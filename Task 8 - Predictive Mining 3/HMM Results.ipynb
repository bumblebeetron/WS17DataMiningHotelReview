```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
oneResults = np.load('OneGrams_TRTE10.npy')
twoResults = np.load('TwoGrams_TRTE10.npy')
thrResults = np.load('ThreeGrams_TRTE10.npy')
```

```python
def plotHist(results, pltName = ''):
    perf = np.zeros(20)
    for i in results:
        if i != -1:
            perf[i] += 1
    perf /= perf.sum()
    perf *= 100
    
    plt.rcParams["figure.figsize"] = (10,5)
    plt.bar(np.arange(20)+1, perf, align='center', alpha=0.5)
    plt.title(pltName)
    plt.xticks(np.arange(20) + 1)
    plt.yticks(np.arange(0,26,5))
    plt.ylabel('Success Percentage')
    plt.xlabel('Accurate Guess')
 
```

```python
plotHist(oneResults, 'Distribution of 1-Grams Accurate Guesses')
plt.savefig('oneResultsDist.png')
plt.show()
```

```python
plotHist(twoResults, 'Distribution of 2-Grams Accurate Guesses')
plt.savefig('twoResultsDist.png')
plt.show()
```

```python
plotHist(thrResults, 'Distribution of 3-Grams Accurate Guesses')
plt.savefig('threeResultsDist.png')
plt.show()
```
