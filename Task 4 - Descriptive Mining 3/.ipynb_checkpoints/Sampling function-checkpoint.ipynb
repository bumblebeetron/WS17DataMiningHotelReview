```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hotelReviews = pd.read_csv('C:\Users\deniz\Documents\Hotel_Reviews.csv')
```

```python
def sampleFromDatFrame(inDataFrame, keyWord, sampleRatio = 0.1, column = 'Hotel_Name'):
    odf = hotelReviews.loc[hotelReviews[column] == keyWord]
    odf = odf.iloc[np.random.choice(odf.shape[0], int(odf.shape[0] * sampleRatio))]
    
    return odf
```
